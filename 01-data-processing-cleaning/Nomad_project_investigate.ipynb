```python
import unicodedata

def normaliser_colonne(nom_colonne):
    # Convertir en minuscules
    nom_colonne = nom_colonne.lower()
    # Remplacer les accents
    nom_colonne = unicodedata.normalize('NFKD', nom_colonne).encode('ASCII', 'ignore').decode('ASCII')
    # Remplacer les caract√®res non alphanum√©riques par des underscores
    nom_colonne = ''.join(char if char.isalnum() or char.isspace() else '_' for char in nom_colonne)
    # Remplacer les espaces multiples par un seul underscore
    nom_colonne = '_'.join(nom_colonne.split())
    return nom_colonne

# Appliquer la normalisation aux noms des colonnes
df.columns = [normaliser_colonne(col) for col in df.columns]

# Afficher les noms de colonnes normalis√©s
print(df.columns)
```

#### 2. Traitement des Valeurs Manquantes

**Imputation des Valeurs Manquantes**

Pour les colonnes num√©riques imputer les valeurs manquantes avec la moyenne ou la m√©diane. Pour les colonnes cat√©gorielles, utiliser la valeur la plus fr√©quente.

Pour les colonnes telles que operation_name, tu peux cr√©er une variable binaire indiquant la pr√©sence ou l'absence d'une op√©ration :

```python
# Cr√©er des variables binaires pour les op√©rations promotionnelles
df['has_operation'] = df['operation_name'].notna().astype(int)
df['has_display_date_gs'] = df['display_date_gs'].notna().astype(int)

# Remplacer les valeurs manquantes dans les colonnes de dates par une valeur neutre (par exemple, 0 ou une date de r√©f√©rence)
df['startdate_op'].fillna(pd.Timestamp('2000-01-01'), inplace=True)
df['enddate_op'].fillna(pd.Timestamp('2000-01-01'), inplace=True)

# Convertir les dates en nombres (jours depuis une date de r√©f√©rence)
date_reference = pd.Timestamp('2000-01-01')

def date_to_numeric(date_series):
    return (date_series - date_reference).dt.days

df['startdate_op_numeric'] = date_to_numeric(df['startdate_op'])
df['enddate_op_numeric'] = date_to_numeric(df['enddate_op'])
```

**Traitement des Colonnes de Date**

Les colonnes de date peuvent √™tre converties en caract√©ristiques num√©riques si elles sont pertinentes pour la r√©gression.

```python
# Convertir les dates en nombres (jours depuis une date de r√©f√©rence)
date_reference = pd.Timestamp('2000-01-01')

def date_to_numeric(date_series):
    return (date_series - date_reference).dt.days

df['order_date_numeric'] = date_to_numeric(df['order_date'])
df['startdate_op_numeric'] = date_to_numeric(df['startdate_op'])
df['enddate_op_numeric'] = date_to_numeric(df['enddate_op'])
df['tracking_day_front_numeric'] = date_to_numeric(df['tracking_day_front'])
```

#### 3. Pr√©paration des Donn√©es

Une fois les valeurs manquantes trait√©es, d√©finis les variables ind√©pendantes et la variable cible pour la r√©gression lin√©aire.



```python
from sklearn.model_selection import train_test_split

# D√©finir les variables ind√©pendantes et la variable cible
features = ['avg_price', 'indice_avg_price', 'display_date_gs_numeric',
             'impression_gs', 'acquisition_cost_gs', 'order_date_numeric',
             'startdate_op_numeric', 'enddate_op_numeric', 'has_operation',
             'has_display_date_gs']

X = df[features]
y = df['total_customers']

# Diviser les donn√©es en ensembles d'entra√Ænement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# V√©rifier les valeurs manquantes dans X_train et X_test
print(X_train.isna().sum())
print(X_test.isna().sum())
```

```python
from sklearn.impute import SimpleImputer

# Cr√©er un imputer pour les colonnes num√©riques
num_imputer = SimpleImputer(strategy='mean')

# Imputer les valeurs manquantes dans X_train et X_test
# Note : Assure-toi que X_train et X_test contiennent les colonnes avec des valeurs manquantes
X_train_imputed = num_imputer.fit_transform(X_train)
X_test_imputed = num_imputer.transform(X_test)

# V√©rifier les valeurs manquantes apr√®s imputation
print(pd.DataFrame(X_train_imputed, columns=X_train.columns).isna().sum())
print(pd.DataFrame(X_test_imputed, columns=X_test.columns).isna().sum())
```

## Mod√®le de donn√©es

#### R√©gression Lin√©aire

Apr√®s avoir pr√©par√© les donn√©es, ajuste le mod√®le de r√©gression lin√©aire et √©value sa performance.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Cr√©er et ajuster le mod√®le de r√©gression lin√©aire
model = LinearRegression()
model.fit(X_train_imputed, y_train)

# Pr√©diction et √©valuation
y_pred = model.predict(X_test_imputed)
print(f'RMSE: {mean_squared_error(y_test, y_pred, squared=False)}')
print(f'R^2: {r2_score(y_test, y_pred)}')

# Coefficients du mod√®le
coefficients = pd.DataFrame({'Feature': X_train.columns, 'Coefficient': model.coef_})
print(coefficients)
```

Les r√©sultats que tu obtiens pour le mod√®le de r√©gression lin√©aire montrent un RMSE relativement √©lev√© et un R2 assez bas, ce qui sugg√®re que le mod√®le a du mal √† expliquer la variance des donn√©es. Voici quelques points √† consid√©rer pour am√©liorer le mod√®le et l'interpr√©ter correctement :

**Analyse des R√©sultats**

_Performance du Mod√®le_ :
- **RMSE (Root Mean Squared Error)** : √Ä 19.26, cela indique l'erreur moyenne que fait le mod√®le lors de la pr√©diction. Plus cette valeur est √©lev√©e, moins le mod√®le est pr√©cis.
- **R2** : √Ä 0.085, cela signifie que le mod√®le explique seulement 8.5% de la variance des donn√©es. Un R2 bas indique que le mod√®le ne capture pas bien les relations entre les variables ind√©pendantes et la variable cible.

**Interpr√©tation des Coefficients**

- Les coefficients montrent l'impact de chaque caract√©ristique sur la variable cible. Les coefficients n√©gatifs et positifs indiquent la direction de l'effet (diminution ou augmentation de la variable cible).
- Par exemple, has_operation a un coefficient tr√®s √©lev√© et n√©gatif, ce qui peut indiquer que les produits en op√©ration promotionnelle ont un effet important sur la variable cible (peut-√™tre en r√©duisant les ventes ou en augmentant les ventes dans une certaine direction).

```python
from sklearn.model_selection import GridSearchCV

# D√©finir les param√®tres √† tester
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Cr√©er un mod√®le de for√™t al√©atoire
rf = RandomForestRegressor(random_state=42)

# D√©finir la recherche par grille
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Ajuster la recherche par grille
grid_search.fit(X_train_imputed, y_train)

# Meilleurs param√®tres
print("Meilleurs param√®tres : ", grid_search.best_params_)

# Meilleur mod√®le
best_rf_model = grid_search.best_estimator_

# √âvaluation
y_pred_best_rf = best_rf_model.predict(X_test_imputed)
print(f'RMSE (Best Random Forest): {mean_squared_error(y_test, y_pred_best_rf, squared=False)}')
print(f'R^2 (Best Random Forest): {r2_score(y_test, y_pred_best_rf)}')
```
